```python
import pandas as pd

import plotly.express as px

import dash
import dash_bootstrap_components as dbc
from dash impo
-+rt dcc, html,dash_table
from dash.dependencies import Input, Output, State
```

```python
df = pd.DataFrame(pd.read_csv("../Data/processeddata.csv"))
```

```python
df
```

```python
app = dash.Dash(external_stylesheets=[dbc.themes.BOOTSTRAP])
app.title = "Movie review analysis"
```

```python
SIDEBAR_STYLE = {
    'position' : 'fixed',
    'top' : '5px',
    'left' : '5px',
    'width' : '25%',
    'padding' : '20px 10px',
    'background-color': '#f8f9fa'
}

HEADER_STYLE = {
    'position' : 'fixed',
    'top' : '5px',
    'margin-left' : '5px',
    'padding' : '20px 10px',
    'background-color' : '#f8f9fa'
}

CONTENT_STYLE = {
    'position' : 'relative',
#    'right' : '5px',
    'margin-left' : '5px',
    'margin-top' : '5px',
    'padding' : '20px 10px',
    'background-color' : '#f8f9fa'
}

TEXT_STYLE = {
    'textAlign': 'center',
    'color': '#191970'
}

GRAPHIC_STYLE = {
    'width' : '90%',
    'margin-left': '10px',
    'margin-right' : '10px',
    'margib-bottom' : '10px'
}
```

```python
sidebar = html.Div(
    [
        html.H2("Sentiment Analysis", style  = TEXT_STYLE),
        html.Hr(),
        dbc.Row([
           dbc.Col([
               html.P('IMDB Movie reviews', style=TEXT_STYLE),
               html.Br(),
               dbc.CardBody([
                   html.Img('source/Capture001.png', style=GRAPHIC_STYLE)
               ])
           ])
        ]),
        dbc.Row([
             dbc.CardBody([
                   html.Img('source/Capture001.png', style=GRAPHIC_STYLE)
               ])
        ]),
        dbc.Row([
            dcc.RadioItems('Positive','Negative')
        ])
    ],
    style = SIDEBAR_STYLE
)
```

```python
header = html.Div([
    dbc.Row([
        dbc.Col([
            dcc.Upload(
                html.A('Choose Dataset')
            )
        ]),
        dbc.Col([
            dbc.Row([
                html.H5('Total number of reviews', style=TEXT_STYLE)
            ]),
            dbc.Row([
                html.P('Positive reviews', style=TEXT_STYLE)
            ]),
            dbc.Row([
                html.P('Negative reviews', style=TEXT_STYLE)
            ])
        ]),
        dbc.Col([
            dbc.Row([
                html.H4('Top Word', style=TEXT_STYLE)
            ]),
            dbc.Row([
                html.P('hello', style=TEXT_STYLE)
            ])
        ])
    ])
],
style = HEADER_STYLE
)
```

```python
pie_chart = px.pie(df, values='sentiment', names='sentiment', title='Distribution of reviews')
```

```python
content = html.Div([
    dbc.Col([
        dbc.Row([
            html.H2('IMDB dataset')
        ]),
        dbc.Row([
            dash_table.DataTable(
            id='my-table',
            columns=[{"name": i, "id": i} for i in df.columns],
            data=df.to_dict('records')   
            )
        ])
    ],
    width = 4
    ),
    dbc.Col([
        dbc.Row([
            dbc.CardBody([
                dcc.Graph(figure = pie_chart, style = GRAPHIC_STYLE)
            ])
        ])
    
    ])
],
style = CONTENT_STYLE
)
```

```python
app.layout = html.Div([
    dbc.Col([
        sidebar
    ], width = 3
    ),
    dbc.Col([
        dbc.Row([header]),
        dbc.Row([content])    
    ])
])
```

```python
if __name__ == '__main__':
    app.run_server()
```
